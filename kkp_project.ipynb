### Let's Dive Into Some More Detailed Analysis! ###

### Before we begin ###

I first want to create a DataFrame that contains everyone who is not an alumni or dropped but is still an active, MC, or conditional member of Kappa Kappa Psi. I am doing this because I want to make sure I have access to data that is current.

*Note: Of course, if later we want to include those categories, we certainly can.*

```python
# This DataFrame includes everyone that is currently an active, MC, or conditional member of Kappa Kappa Psi
df_not_alumn_or_drop = df[((df.Alumn == "No") & (df.Active == "Yes")) | (df.Conditional == "Yes") | (df.MC == "Yes")].reset_index()
df_not_alumn_or_drop
```

Kappa Kappa Psi is described as a music service fraternity, so you would expect there to be current brothers who might be pursuing music as a career. How might we figure out what that percentage is? With this DataFrame, we are able to determine that information!

```python
# This DataFrame includes everyone in Kappa Kappa Psi that is currently pursuing music
df_music = df_not_alumn_or_drop[(df_not_alumn_or_drop["Major"].str.contains("Music")) | (df_not_alumn_or_drop["Major"].str.contains("Conducting")) | (df_not_alumn_or_drop["Major"].str.contains("Performance"))].reset_index()
df_music
```

```python
# This DataFrame represents the percentage of Music Majors in Kappa Kappa Psi
df_music_percentage = df_music["First Name"].count() / df_not_alumn_or_drop["First Name"].count() * 100
df_music_percentage
```

Well, would you look at that! Around 51.28% of the `df_not_alumn_or_drop` DataFrame are pursuing music right now! That's over half of our current brothers!

But what about the other majors besides music? Well, with the `df_not_alumn_or_drop` DataFrame, we can actually figure out all of the majors that are represented in Kappa Kappa Psi right now!

Let's take a look at all of these majors!

```python
# This DataFrame includes everyone's majors in Kappa Kappa Psi as well as how many people represent each major.
df_major_count = df_not_alumn_or_drop.groupby("Major").agg("count").reset_index()
df_major_count [["First Name", "Major"]] 
```

```python
# This DataFrame caluclates the total number of majors that are currently represented in Kappa Kappa Psi
df_number_of_majors = df_major_count["Major"].count()
df_number_of_majors
```

Wow! Kappa Kappa Psi is currently representing 20 different majors! That is a lot of diversity!

I think it is important that as a fraternity, we have an idea of what our growth looks like each year. To do that, we can look at the percentage of people in each year of graduation. I hope to eventually add a column for the specific pledge classes since I know not everyone rushed Kappa Kappa Psi their freshman year; however, until then, the estimated year of graduation can work in the meantime.

```python
# This DataFrame includes the total number of people that graduate in the same year as well as the percentage
# This DataFrame includes every single year inlcluding the alumns
df_year_all = df.groupby("Expected Graduation Year").agg("count").reset_index()
df_name_sum = df_year_all["First Name"].sum()
df_year_all["Year Percentage"] = df_year_all["First Name"] / df_name_sum * 100
df_year_all [["Expected Graduation Year", "First Name", "Year Percentage"]]
```

```python
# This DataFrame includes the total number of people that graduate in the same year as well as the percentage
# This DataFrame only includes the current members
df_year_now = df_not_alumn_or_drop.groupby("Expected Graduation Year").agg("count").reset_index()
df_name_sum = df_year_now["First Name"].sum()
df_year_now["Year Percentage"] = df_year_now["First Name"] / df_name_sum * 100
df_year_now [["Expected Graduation Year", "First Name", "Year Percentage"]]
```

I included two versions of the DataFrame, one for only current members and one for all of the data so we can observe the growth as a whole, as well as our current representation and growth.

Sometimes, data can be challenging to interpret when reading it as a table. However, it can become much easier to understand when presented visually. Below are two bar graph visualizations for the `df_year_all` and `df_year_now` DataFrames above.

```python
#This bar graph inclues every single year
df_year_all.plot.bar(x='Expected Graduation Year', y='Year Percentage')
```

```python
#This bar graph inclues all of the current years
df_year_now.plot.bar(x='Expected Graduation Year', y='Year Percentage')
```
